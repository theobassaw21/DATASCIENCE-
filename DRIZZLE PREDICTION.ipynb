```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import missingno as msno
#pd.set_option('display.max_rows')
plt.style.use('dark_background')
```

```python
data = pd.read_csv('file.csv')
data
```

```python
data.head(10)
```

```python
data.drop('Date/Time',axis=1,inplace=True)
```

```python
data
```

```python
msno.bar(data)
```

```python
data=data.join(pd.get_dummies(data.Weather))
data
```

```python
data=data[['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%', 'Wind Speed_km/h',
       'Visibility_km', 'Press_kPa', 'Drizzle']].copy()
```

```python
data
```

```python
x = data.drop(columns='Drizzle')
```

```python
y = data.Drizzle
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
x_train
```

```python
y_train.values.reshape(-1,1)
```

```python
y_pre_train=lr.predict(x_train)
```

```python
lr.score(x_train,y_train)
```

```python
plt.scatter(y_train,y_pre_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
ypred_t=model.predict(x_train)
```

```python
model.score(x_train,y_train)
```

```python
model.fit(x_test,y_test)
```

```python
ypred_te=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix , accuracy_score
```

```python
print(confusion_matrix(y_test,ypred_te))
print(classification_report(y_test,ypred_te))
print(accuracy_score(y_test,ypred_te))
```
